```python
import pandas as pd #digunakan untuk manipulasi data
import numpy as np #digunakan untuk operasi numeris
import matplotlib.pyplot as plt #digunakan untuk visualisasi data
import seaborn as sns #digunakan untuk visualisasi data
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Memuat dataset dan menyimpannya pada dataframe
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Amikom Sehat/stroke.csv')
data
```

```python
#menghapus kolom id
data = data.drop(columns='id', axis=1)
data
```

```python
#mengetahui jumlah baris dan kolom
data.shape
```

```python
#mengetahui info data
data.info()
```

```python
#cek apakah ada data kosong
data.isnull().sum()
```

```python
#melihat statistik data
data.describe()
```

```python
# memeriksa distribusi data Variabel Stroke ( 1 : Stroke , 0 : Sehat )
data['stroke'].value_counts()
```

```python
#X : data selain kolom stroke
X = data.drop(columns='stroke', axis=1)

#Y : data kolom stroke
Y = data['stroke']
```

```python
X
```

```python
Y
```

```python
#memisahkan data mejadi data training dan data test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=2)
```

```python
#menampilkan data
print(X.shape, X_train.shape, X_test.shape, Y.shape, Y_train.shape, Y_test.shape)
```

```python
#Menggunakan model logika regressi
model = LogisticRegression()
```

```python
#Training model logika regressi dengan data training
model.fit(X_train, Y_train)
```

```python
# Acurracy pada data training

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
input_data = (0,74,0,0,1,1,74.96,26.6)

#mennganti input data menjadi numpy array
input_data_array = np.array(input_data)

#
input_data_reshape = input_data_array.reshape(1,-1)

prediksi = model.predict(input_data_reshape)
print(prediksi)

if (prediksi[0]==0):
  print('Anda tidak mempunyai penyakit Stroke')
else:
  print('Anda mempunyai penyakit Stroke')
```
