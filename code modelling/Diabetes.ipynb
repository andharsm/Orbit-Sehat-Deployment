```python
#Import library
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dd = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Amikom Sehat/Data_diabetes (1).csv', 
                 usecols= ['Sex', 'Age',    'Diabetes_012', 'HighBP',   'HighChol',    'BMI',  'Smoker',   'Stroke',   
                 'HeartDiseaseorAttack',    'PhysActivity','Fruits', 'Veggies'])
```

```python
#X : data selain kolom target
X = dd.drop(columns='Diabetes_012', axis=1)

#Y : data kolom target
Y = dd['Diabetes_012']
```

```python
#melihat data x
X
```

```python
#melihat data Y
Y
```

```python
#memisahkan data mejadi data training dan data test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```



---

**MODEL TRAINING**

```python
#Menggunakan model logika regressi
model = LogisticRegression()
```

```python
#Training model logika regressi dengan data training
model.fit(X_train, Y_train)
```



---

**MODEL EVALUASI**

```python
# Acurracy pada data training

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy pada data training : ', training_data_accuracy)
```

```python
import pickle
filename = 'model_Diabetes.pkl'
pickle.dump(model, open(filename, 'wb'))
```

**MEMBUAT SYSTEM PREDIKSI**


```python
#input : Jenis Kelamin, Umur, Tekanan darah, colestroll, BMI, Merokok,  Stroke, Serangan jantung, Aktivitas berat, Mengkonsumsi buah, Mengkonsumsi sayur.
input_data = (1,	30,	1,	1,	46,	1,	1, 0,	1,	1,	0)

#mennganti input data menjadi numpy array
input_data_array = np.array(input_data)

#
input_data_reshape = input_data_array.reshape(1,-1)

prediksi = model.predict(input_data_reshape)
print(prediksi)

if (prediksi[0]==0):
  print('Anda tidak terindikasi penyakit Diabetes')
elif (prediksi[0]==1):
  print('Anda terindikasi prediabetes')
else:
  print('Anda terindikasi penyakit Diabetes')
```
