```python
#import dependency
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#import dataset dari gdrive
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Amikom Sehat/bmi.csv')
```

```python
#cek distribusi data variabel target 0 = sangat kurus, 1 = kurus, 2 = normal, 3 = kegemukan, 4 = obesitas, 5 = obesitas ekstrim
data['Index'].value_counts()
data['Gender'].value_counts()
```

```python
#pembagian data
#X = data selain Index yang dicari
x = data.drop(columns='Index', axis=1)

#y = data Index / yang dicari
y = data['Index']
```

```python
#lihat data x
x
```

```python
#lihat data y
y
```

```python
#membagi data training dan data test
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,stratify=y, random_state=2)
```

```python
#cek hasil pembagian data training dan data test
print(x.shape, x_train.shape, x_test.shape, y.shape, y_train.shape, y_test.shape)
```

```python
#Model Data
#Logika Regresi
model=LogisticRegression()
```

```python
#Training model logika regressi dengan data training
model.fit(x_train, y_train)
```

```python
#Evaluasi
# Acurracy pada data training

x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)

print('Accuracy pada data training : ', training_data_accuracy)
```

```python
import pickle
filename = 'model_BMI.pkl'
pickle.dump(model, open(filename, 'wb'))
```

```python
#input(Gender, Tinggi, Berat)
input_data = (0,180,80)

#mennganti input data menjadi numpy array
input_data_array = np.array(input_data)

#
input_data_reshape = input_data_array.reshape(1,-1)

prediksi = model.predict(input_data_reshape)
print(prediksi)

if (prediksi[0]==0):
  print('Anda Sangat Kurus')
elif (prediksi[0]==1):
  print('Anda Kurus')
elif (prediksi[0]==2):
  print('Anda Normal')
elif (prediksi[0]==3):
  print('Anda Kegemukan')
elif (prediksi[0]==4):
  print('Anda Obesitas')
else:
  print('Anda Obesitas Ekstrim')
```
